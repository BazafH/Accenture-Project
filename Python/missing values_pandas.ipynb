```python
import numpy as np
import pandas as pd
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df=pd.DataFrame(people)
```

```python
df
```

```python
df.dropna()                                                #working on numpy and pandas none values/empty values
```

```python
df.dropna(axis='index', how='any')                        #default set parameters
```

```python
df.dropna(axis='columns', how='any')  #removed columns that contain any none/np.nan values
```

```python
df.dropna(axis='index', how='any', subset=['age'])     #for single subset value, how argument is irrelevant. Subset means to include col name so that
                                                        # so that it will look for col values. In case of this example, it will look for 'age' values
                                                        # and eliminated those rows which have empty(none,nan) age values. 
    
```

```python
df.dropna(axis='index', how='all', subset=['last','email'])
```

```python
df.dropna(axis='index', how='any', subset=['last','email'])
```

```python
df.isna()   #give dataframe with boolean values. Empty(none/np.nan) values are true, remaining all are false.
```

```python
df['age'].fillna(0, inplace=True)         #Empty(none/np.nan) values are filled with the given value. Helpful in numerical data
```

```python
df
```

```python
df.replace({'NA':None, 'Missing':None},inplace=True)            # Wrong data can be manipulated using replace.
                                                                # Useful for cleaning data
df
```

```python
df.dropna()
```
